# **régression linéaire**


**La régression linéaire ? De quoi parle t-on ?**
- Certaines personnes aiment donner des noms compliqués pour des choses intuitives à comprendre.
- La régression linéaire en est un bon exemple.
- derrière ce nom, se cache un concept très simple : La régression linéaire est un algorithme qui va trouver une droite qui se rapproche le plus possible d’un ensemble de points.
- Les points représentent les données d’entraînement (Training Set).

- Schématiquement, on veut un résultat comme celui là :


image.png



- Nos points en orange sont les données d’entrée (input data). Ils sont représentés par le couple (x_{i}, y_{i}). Les valeurs x_{i} sont les variables prédictives, et y_{i} est la valeur observée (le prix d’une maison par exemple).
.
- On cherche à trouver une droite F(x) = \alpha*x + \beta tel que, quelque soit x_{i}, on veut que F(x_{i}) \approx y_{i}.

- En d’autres termes, on veut une droite qui soit le plus proche possible de tous les points de nos données d’apprentissage.



```python
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# define the x and y arrays
# data with medium fit
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [98,85,86,89,112,85,101,84,93,79,76,82,81]

'''
#data with best fit
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [14,35,48,36,1,254,2,63,9,99,120,62,25]
'''

'''
#data with worst fit
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [63,9,99,62,25,14,35,48,36,1,254,2,63]
'''
```

```python
#calculating linear regression using linregress() function
slope, intercept, r, p , std_error = stats.linregress(age_x,capacity_y)
```

```python
def find_new_y(x):
    #find the value of y = mx+b from the given x value
    y = slope*x + intercept
    return y
```

```python
# for every value of x in the array, find the new y value and save as a list
new_y_values = list(map(find_new_y, age_x))
```

```python
#plotting the original data as scatter plot
#plotting the line of regression
#draw the scatter plot
plt.scatter(age_x, capacity_y, label='original data')
plt.plot(age_x, new_y_values, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
print("r value = "+str(r))
```

```python
#predict the production capacity for a 15 year old machine
predicted_capacity_y = find_new_y(15)
print("predicted capacity is "+str(predicted_capacity_y))
```
